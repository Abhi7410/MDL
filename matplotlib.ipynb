```python

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style 

xpoints = np.array([1,2,5,6,8])
ypoints = np.array([3, 10,19,10,2])
style.use("ggplot")
# for grid
plt.plot(xpoints, ypoints)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("first",fontdict = font1,loc = 'left')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style 

xpoints = np.array([1,2,5,6,8])
ypoints = np.array([3, 10,19,10,2])
style.use("ggplot")
# for grid
# use markers for vertices marker="o"
plt.plot(xpoints, ypoints,'d--c',ms=20,mec='r',mfc='k')
# ms for size of markers 
# mec for marker color boundary
# mfc for marker face color (density)
# --r is for red and , b for blue,y for yellow
# r = red
# g = green
# b = blue
# c = cyan
# m = magenta
# y = yellow
# k = black
# w = white
# d = diamond , h = hexagon, p = pentaogn,v = triangle down,...
# ':' doted line , '-' solid line, '-.' dashed line /dotted, '--' dashed
plt.title("first")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# line style
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted',color ='k',linewidth='10')
plt.plot(ypoints,color='r',linewidth='5')

# dotted,dashed
# line width for bold line r more width
plt.show()
```

```python
# With the subplots() function you can draw multiple plots in one figure:
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 3, 1)
plt.title("Mitul")
plt.plot(x,y)
#the figure has 1 row, 2 columns, and this plot is the second plot.
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 3, 2)
plt.title("Sarthak")
plt.plot(x,y)

x = np.array([0, 1, 2, 3,10])
y = np.array([10, 20, 30, 0,10])

plt.subplot(1, 3, 3 )
plt.title("Tarun")
plt.plot(x,y)

plt.suptitle("MY SHOP")
plt.show()
```

```python
# bar graph
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,color='k')
plt.show()
```

```python
# pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, -0.4, 0.1, 0.2]
plt.pie(y,labels=mylabels,startangle = 60,explode=myexplode,shadow=True)
plt.show() 
```

```python
# pie chart
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y,labels=mylabels,startangle = 60,shadow=True)
plt.legend()
# to show separately
plt.show() 
```
